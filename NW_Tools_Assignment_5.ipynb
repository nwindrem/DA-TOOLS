Define Visualizations

```python
#define visualization function
def visualize(filename):
    
    "load data without headers"
    data = pd.read_csv(filepath_or_buffer = filename,header=None)
    
    "Find max value per column"
    col_max = data.max(axis = 0)
    
    "find min value per column"
    col_min = data.min(axis = 0)
    
    "calculate average value per column"
    col_avg = data.mean(axis = 0)
    
    "Create subplot"
    fig1=plt.subplot()

    plt.plot (col_max, label = "Max")
    plt.plot(col_avg, label = "Mean")
    plt.plot (col_min, label = "Min")
    
    "label plot"
    plt.legend()
    plt.title('MIN, MEAN and MAX')
    plt.suptitle (filename)
    plt.xlabel ("Features")
    plt.ylabel ('Frequency')
    
    "Display the plot"
    matplotlib.pyplot.show()
```

run viz and error functions for all 4 datasets

```python
#run visualization and error functions for all 4 datasets 
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    err(datasets)
```
